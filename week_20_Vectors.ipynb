```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams['font.size']=16
v=np.array ([2,3]);
v0=3 * v
plt.plot((0,v0[0]),(0,v0[1]),'r', \
         label=r'$3 \cdot v$')
plt.plot ((0,v[0]),(0,v[1]),'b',\
          label=r'$v$')
plt.legend()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams['font.size']=16
v=np.array ([1,1]);
v0=np.sqrt(1/2) * v
plt.plot((0,v0[0]),(0,v0[1]),'r', \
         label=r'$3 \cdot v$')

plt.plot ((0,v[0]),(0,v[1]),'b',\
          label=r'$v$')

print(np.linalg.norm(v0))
plt.legend()
```
