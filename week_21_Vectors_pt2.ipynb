```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#vector X itself = by its transpose/ v^T
 #||v|| = length
  #v.v^T = ||v||^2
```

```python
#||v|| = length
```

```python
v=np.array([3,3])
u=np.array([2,2])
d=u-v
plt.plot([0,v[0]],[0,v[1]])
plt.plot([0,u[0]],[0,u[1]])
plt.plot([v[0],u[0]],[v[1],u[1]])
distance=np.linalg.norm(d)
print(distance)
print(np.sqrt (d@d))
print (d)
```
